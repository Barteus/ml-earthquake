```python
import pandas as pd
import numpy as np
import seaborn as sns

import sys
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
np.set_printoptions(threshold=sys.maxsize)

%matplotlib inline 
```

```python
train_values = pd.read_csv('data/train_values.csv', index_col='building_id')
train_labels = pd.read_csv('data/train_labels.csv', index_col='building_id')
train = train_values.merge(train_labels, left_index=True, right_index=True)
train.sort_index(inplace=True)
train.head()
```

```python
train.groupby(by=['damage_grade']).count().iloc[:,0]
```

```python
train.info()
```

```python
train.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_enc = train.copy()
for col in train.select_dtypes(object).columns:
    df_enc[col] = LabelEncoder().fit_transform(train[col].values.reshape(-1,1))
```
